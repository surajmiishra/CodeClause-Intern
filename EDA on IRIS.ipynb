# Load the Dataset

```python
import pandas as pd

# Load the dataset
data_path = 'IRIS.csv'
df = pd.read_csv(data_path)

# Display the first few rows of the dataframe
df.head()
```

# Understand the Dataset

```python
# Display basic information about the dataset
df.info()

# Display summary statistics
df.describe()

# Check for missing values
print(df.isnull().sum())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

```

# Data Visualization

```python
# Visualize Distributions of Features

import matplotlib.pyplot as plt
import seaborn as sns

# Set the style of the visualization
sns.set(style="whitegrid")

# Plot histograms for each feature
df.hist(bins=20, figsize=(10, 8), grid=False)
plt.tight_layout()
plt.show()
```

```python
# Pairplot to visualize relationships between features
sns.pairplot(df, hue='species', markers=["o", "s", "D"])
plt.show()

# Correlation heatmap (excluding the 'species' column)
plt.figure(figsize=(8, 6))
sns.heatmap(df.drop('species', axis=1).corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.show()
```

```python
# Box Plots and Violin Plots

# Box plots
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, orient="h")
plt.title('Box Plot of All Features')
plt.show()

# Violin plots
plt.figure(figsize=(12, 8))
sns.violinplot(x='species', y='sepal_length', data=df)
plt.title('Violin Plot of Sepal Length by Species')
plt.show()

plt.figure(figsize=(12, 8))
sns.violinplot(x='species', y='sepal_width', data=df)
plt.title('Violin Plot of Sepal Width by Species')
plt.show()

plt.figure(figsize=(12, 8))
sns.violinplot(x='species', y='petal_length', data=df)
plt.title('Violin Plot of Petal Length by Species')
plt.show()

plt.figure(figsize=(12, 8))
sns.violinplot(x='species', y='petal_width', data=df)
plt.title('Violin Plot of Petal Width by Species')
plt.show()
```

# Conclusions


After visualizing the data, we can summarize our findings.

Distribution of Features: Histograms show the distribution of each feature.
Relationships Between Features: Pair plots show the relationships between features, colored by species. For example, petal length and petal width seem to have a strong correlation.
Species Differences: Violin plots and box plots highlight the differences in feature distributions across different species.
